Demand function: q=ap^e where a is a constant < 0 & e < 0 is elasticity of demand

Supply function: q=bp^n where b is a constant < 0 & n < 0 is elasticity of supply

q = q  so   ap^e = bp^n     ap^e-n = b

p = (b/a)^1/e-n

q=a[(b/a)^1/e-n]^e

```python
#a=.5 b=2 e=-2 n=.5
a, b, e, n = 0.5, 2, -2, 0.5
#Initial guess
pnew = 1
#Apply the algorithm repeatedly for at most 100 times
for i in range(100):
    #Update "recursively" the price (i.e., change it from the old to the new)
    p=pnew
    #fuction: 0 = ap^e - bp^n
    f=a * p **e - b * p **n
    #Compute the derivative of the function whose zero we are trying to find
    d=a * e * p**(e - 1) - b * n * p**(n - 1)
    #Compute the updating factor
    deltap = -f/d
    #Compute the update
    pnew = p + deltap
    #Display each itermediate step
    print('The current iteration is',i)
    print('The current guess for the price is {:0.10f} '.format(p))
    print('The current value for f is {:0.10f} '.format(f))
    print('The current value for d is {:0.10f} '.format(d))
    print('The current value for the update is {:0.10f} '.format(deltap))
    print('_____________________________')
    #Check if the update satisfies our convergence criterion
    if abs(deltap) < 1.e-8: #  accuracy
        break
#Format and display the final output        
print('The market clearing price is {:0.3f} '.format(p))
print('The number of iterations is',i)
print('The last update is {:0.10f} '.format(deltap))
```

adding a do-loop for multiple e values, where at each iteration the excess demand function is plotted.

```python
import matplotlib.pyplot as plt
# Given constants
e_values = [-2.0, -1.0]
a, b, n = 0.5, 2, 1
# Initial guess
pnew = 1
# Apply the algorithm repeatedly for at most 100 times
for e in e_values:
    # Lists to store values for plotting
    xvalues = []
    yvalues = []
    for i in range(100):
        # Update "recursively" the price (i.e., change it from the old to the new)
        p = pnew
        # Excess demand function: 0 = ap^e - bp^n
        f = a * p ** e - b * p ** n
        # Compute the derivative of the function
        d = a * e * p**(e - 1) - b * n * p**(n - 1)
        # Compute the updating factor
        deltap = -f / d
        # Compute the update
        pnew = p + deltap
        # Store values for plotting
        xvalues.append(i)
        yvalues.append(f)
        # Check if the update satisfies our convergence criterion
        if abs(deltap) < 1.e-8:  # accuracy
            break
    # Format and display the final output
    print('The market clearing price for e = {} is {:0.3f}'.format(e, p))
    print('The number of iterations is', i)
    print('The last update is {:0.10f} '.format(deltap))
    # Plot the excess demand function at each iteration for the current e value
    plt.plot(xvalues, yvalues, marker='o', label=f'e = {e}')
# Display the plot
plt.title('Excess Demand Function at Each Iteration for Different e Values')
plt.xlabel('Iteration')
plt.ylabel('Excess Demand')
plt.legend()
plt.show()
```
